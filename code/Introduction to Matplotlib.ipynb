```python
#--------------------------------------------------------------------------------
#%%: 第一個圖形，plot()預設動作為在點之間繪製線條

import matplotlib.pyplot as plt
import numpy as np

# 產生X/Y資料
xpoints = np.array([0, 6])    # 0,6
ypoints = np.array([0, 250])  # 0,250

# 顯示X/Y資料: 注意須將X放一起，Y放一起
print(xpoints, ypoints)

# 將X/Y繪製線於座標軸: 從(0,0)到(6,250)
plt.plot(xpoints, ypoints)

# 顯示圖形
plt.show()
```

## 強調標示(Marker)

```python
# https://www.w3schools.com/python/matplotlib_markers.asp
```

```python
#--------------------------------------------------------------------------------
#%%: 

import matplotlib.pyplot as plt
import numpy as np

# 產生Y資料，X則使用預設值
ypoints = np.array([3, 8, 1, 10])

# 可參考: https://matplotlib.org/stable/api/markers_api.html
# 找出不同圖案，可嘗試：P, x, s, *,...
plt.plot(ypoints, marker = 'P')

# 顯示圖形
plt.show()
```

```python
#--------------------------------------------------------------------------------
#%%: fmt: 格式化強調標示

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

# 可參考： https://www.w3schools.com/python/matplotlib_markers.asp
# 找出不同設定

# X使用預設值
plt.plot(ypoints, 'o:r') # o:r意義 -> marker|line|color

# 顯示圖形
plt.show()

# Marker可參考: https://matplotlib.org/stable/api/markers_api.html

# Line reference (線格式)
# '-' > solid line
# ':' > dotted line
# '--' > Dashed line
# '-.' > solid line

# color reference (顏色)
# 'r' > Red
# 'g' > Green
# 'b' > Blue
# 'c' > Cyan
# 'm' > Magenta
# 'y' > Yellow
# 'k' > Black
# 'w' > White

# 請自行測試其他組合
```

```python
#--------------------------------------------------------------------------------
#%%: 強調標示大小

import matplotlib.pyplot as plt
import numpy as np

# 產生Y值，X使用預設值
ypoints = np.array([3, 8, 1, 10])

# 20 points
plt.plot(ypoints, marker = 'o', ms = 20)

# 顯示圖形
plt.show()
```

```python
#--------------------------------------------------------------------------------
#%%: 強調標示外圍顏色(mec)

import matplotlib.pyplot as plt
import numpy as np

# 產生Y值，X使用預設值 
ypoints = np.array([3, 8, 1, 10])

# 20 points, 紅色外圈
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')

# 顯示圖形
plt.show()
```

```python
#--------------------------------------------------------------------------------
#%%: 強調標示內顏色(mfc)

import matplotlib.pyplot as plt
import numpy as np

# 產生Y值，X使用預設值 
ypoints = np.array([3, 8, 1, 10])

# 20 points, 紅色內圈
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')

# 顯示圖形
plt.show()

# 請自行嘗試變更不同組合
```

```python
#--------------------------------------------------------------------------------
#%%: 設定XY軸標題與圖標題字型/大小/顏色

import matplotlib.pyplot as plt
import numpy as np

# 產生XY資料
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# 繪製圖案
plt.plot(x, y)

# 設定字型相關特性: family指字體，color指顏色，size指大小
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

# 設定圖標題
plt.title("Sports Watch Data", fontdict = font1)

# 設定XY軸名稱
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

# 顯示圖案
plt.show()
```

## 同時繪製多張圖(Subplot)

```python
#--------------------------------------------------------------------------------
#%%: subplot設定多張圖

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 設定顯示在第一張次圖
plt.subplot(1, 2, 1) # 第一個1表1列，2表2欄，所以畫布為1*2格式，可容納兩張圖

# ----------------
# -      -       -
# -  1   -   2   -
# -      -       -
# ----------------

# 繪製圖案
plt.plot(x, y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 設定顯示在第二張次圖
plt.subplot(1, 2, 2)

# 繪製圖案
plt.plot(x, y)

# 顯示圖案
plt.show()
```

```python
#--------------------------------------------------------------------------------
#%%: 

import matplotlib.pyplot as plt
import numpy as np


#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 2列1欄，繪製於第1個次圖
plt.subplot(2, 1, 1)

# 繪製圖案
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 2列1欄，繪製於第2個次圖
plt.subplot(2, 1, 2)

# 繪製圖案
plt.plot(x,y)

# 顯示圖案
plt.show()
```

### 繪製兩條線

```python
# 繪製兩條線

# 匯入較常用的繪圖功能
from matplotlib import pyplot as plt

# 設定X與Y值
x1, y1 = [1600, 1700, 1800, 1900, 2000], [0.2, 0.5, 1.1, 2.2, 7.7]
x2, y2 = [1600, 1700, 1800, 1900, 2000], [1, 1, 2, 3, 4]

# 繪製線`:格式可參考https://matplotlib.org/3.3.4/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(x1, y1, 'rx-', label = 'Actual')  # 顏色(紅):標記形式(x標記):線形式(實心)
plt.plot(x2, y2, 'bo--', label = 'Fake')   # 顏色(藍):標記形式(圓形標記):線形式(虛線)

# 標題
plt.title('World Population Over Time')

# X軸
plt.xlabel('Year')

# Y軸
plt.ylabel('Population (billions)')

# 圖例
plt.legend()

# 顯示畫面
plt.show()
```

## 散佈圖(Scatter plot)

```python
# 散步圖用於呈現兩個變數之間的關係，因此兩個變數的個數需相同(陣列大小需相同)，一個變數代表X，另一個變數代表Y
```

### 簡易散佈圖

```python
#--------------------------------------------------------------------------------
#%%: 簡易散佈圖

import matplotlib.pyplot as plt
import numpy as np

# 宣告X/Y變數
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# 使用scatter函數，傳入X/Y
plt.scatter(x, y)

# 顯示圖案: 由圖來看，X與Y成反方向關係
plt.show()
```

### 比較兩個圖

```python
# creating a scatter plot to represent height-weight distribution
# 匯入較常用的繪圖功能
from matplotlib import pyplot as plt
from random import randint

# 設定隨機種子
random.seed(2)

# 在58-78英吋間，產生20筆隨機數字
height = [randint(58, 78) for x in range(20) ]   # 20 records between 4'10" and 6'8"

# 在90-250磅間，產生20筆隨機數字
weight = [randint(90, 250) for x in range(20) ]  # 20 records between 90lbs. and 250lbs.

# 在18-65歲間，產生20筆隨機數字
age = [ randint(18, 65) for x in range(20) ]      # 20 records between 18 and 65 years old

# 散佈圖: https://matplotlib.org/3.3.4/api/_as_gen/matplotlib.pyplot.scatter.html
plt.scatter(weight, height, c = age) # c表color, 以age呈現不同顏色

# 圖標題
plt.title("Height-Weight Distribution")

# X軸標題
plt.xlabel("Weight (lbs)")

# Y軸標題
plt.ylabel("Height (inches)")

# 色條，名稱為Age
plt.colorbar(label = "Age")      # adds color bar to right side

# 顯示圖形
plt.show()
```

```python
# 匯入較常用的繪圖功能
from matplotlib import pyplot as plt
import numpy as np

# 以numpy產生隨機種子
np.random.seed(5)

# 以40為平均數，10為標準差，產生1000個隨機常態分配數字
ages = [np.random.normal(loc = 40, scale = 10) for x in range(1000) ]    # ages centrally distributed around 40

# 顯示前10個年齡
print(ages[0:10])

# 繪製年齡直方圖，切成45個間隔
plt.hist(ages, bins = 45)

# 圖標題
plt.title("Ages per Population")

# X軸名稱
plt.xlabel("Age")

# Y軸名稱
plt.ylabel("# of People")

# 儲存圖檔
plt.savefig('figure10-6.jpg')

# 顯示圖檔
plt.show()
```

### 進階範例

```python
# 匯入較常用的繪圖功能
from matplotlib import pyplot as plt
import numpy as np

# 以60為平均數，15為標準差，產生1000個隨機常態分配數字
florida = [np.random.normal(loc = 60, scale = 15) for x in range(1000)]

# 以35為平均數，5為標準差，產生1000個隨機常態分配數字
california = [np.random.normal(loc = 35, scale = 5) for x in range(1000)]

# chart 1
plt.hist(florida, bins = 45, color = 'r', alpha = 0.4) # 45個間隔，紅色，alpha為透明度，數字越低越淡
plt.show()

# chart 2
plt.hist(california, bins = 45, color = 'b', alpha = 0.2) # 45個間隔，藍色，alpha為透明度，數字越低越淡
plt.show()


# chart 3
plt.hist(florida, bins = 45, color = 'r', alpha = 0.4) # 45個間隔，紅色，alpha為透明度，數字越低越淡
plt.hist(california, bins = 45, color = 'b', alpha = 0.4) # 45個間隔，藍色，alpha為透明度，數字越低越淡

# 儲存圖檔
plt.savefig('figure10-7.jpg')

# 顯示圖檔
plt.show()
```

## 盒形圖(BOX plot)

```python
# 繪製男性/女性身高盒形圖

# 匯入較常用的繪圖功能
from matplotlib import pyplot as plt
import numpy as np

# 男性身高
males = [72, 68, 65, 77, 73, 71, 69]

# 女性身高
females = [60, 65, 68, 61, 63, 64]

# 組合男性與女性身高
heights = np.array([males, females], dtype = object)
# heights = ([males, females]) # 亦可，但有警告訊息

# 調整圖尺寸時使用，未設定則為預設大小
plt.figure(figsize = (15, 8))

# 繪製圖形
plt.boxplot(heights)

# 設定X軸標籤，X軸1的位置是Male，2的位置是Female
plt.xticks([1, 2], ["Male", "Female"])

# 圖標題及字型大小
plt.title("Height by Gender", fontsize = 22)

# X軸標題及字型大小
plt.xlabel('Gender', fontsize = 14)

# Y軸標題及字型大小
plt.ylabel("Height (inches)", fontsize = 14)

# 顯示圖形
plt.show()
```
