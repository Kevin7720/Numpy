### 格式化字串

## .format()函數

```python
# .format()函數：使用大括號{}與點.
name = 'John'

# {}用來設定變數顯示的位置，format()括號裡面的值
print('Hello {}'.format(name))
```

```python
# .format()函數：大括號{}與點.
name = 'John'

# 兩個變數個格式化
print('Hello {}, you are {} years old.'.format(name, 38))
```

### 另一種format方法

```python
# .format()函數：大括號{}與點.

# 另一種方法，同樣結果
name = 'John'

# 將格式指定給txt
txt = 'Hello {}, you are {} years old.'

# 使用format格式化txt，須格式化的資料同樣置於format
print(txt.format(name, 38))
```

### format: 指定位置方法

```python
# 指定位置方法

# 宣告三個變數
quantity = 3     
itemno = 567     # 第1個
price = 49.95    # 第2個

# 格式，並指定變數顯示的項目
myorder = "I want to pay {2} dollars for {0} pieces of item {1}."

# 顯示格式化結果，變數顯示順序：012
print(myorder.format(quantity, itemno, price))

# 第0個：quantity
# 第1個： itemno
# 第2個： price
```

### 指定位置方法，並加上格式化數字

```python
# 指定位置方法，並加上格式化數字

# 宣告三個變數
quantity = 3
itemno = 567
price = 49  

# 格式，並指定變數顯示的項目及小數位數格式化，冒號後的.2f表示小數兩位浮點數，且前面有:
myorder = "I want to pay {2:.2f} dollars for {0} pieces of item {1}."

# 顯示格式化結果
print(myorder.format(quantity, itemno, price))

# 第0個：quantity
# 第1個： itemno
# 第2個： price
```

### 指定索引名稱(Named index) 

```python
# 指定索引名稱(Named index) 

# 宣告格式：大括號{}中即註明變數名稱
myorder = "I want to pay {price} dollars for {quantity} pieces of item {itemno}."

# 顯示格式化結果：直接將變數名稱指定值
print(myorder.format(quantity = 3, itemno = 567, price = 49.95))
```

## 全域變數(Global variable): 整個程式都可使用到的變數

### 宣告於函數外的變數為全域變數 

```python
# 宣告於函數外的變數為全域變數 
x = "awesome" # x為全域變數

# 定義函數，注意冒號，以及下一行需縮排，沒有縮排的地方就非屬於該函數內容
def myfunc():
  print("Python is " + x) # 使用全域變數<---函數到此為止

# 呼叫函數，x並未在myfunc中宣告，但仍可使用
myfunc()
```

### 全域變數與函數內變數相同名稱

```python
# 全域變數與函數內變數相同名稱

x = "awesome" # 全域變數

# 定義函數
def myfunc():
  
  # 宣告與全域變數相同名稱變數
  x = "fantastic"

  print("Python is " + x) # 此處x為函數內的x

# 呼叫函數    
myfunc()

# 此處為仍全域變數
print("Python is " + x)
```

### global關鍵字: 可將函數內變數變為全域變數

```python
# global關鍵字: 可將函數內變數變為全域變數

# 需告函數
def myfunc():
  
  # 變成全域變數
  global x
  
  # 全域變數  
  x = "fantastic"

# 呼叫函數    
myfunc()

# 顯示變數，雖在函數內宣告，但因global指令，也可以變成全域變數
print("Python is " + x)
```

```python
# 請試著思考看看底下print指令的x會顯示哪個值？之後再執行驗證答案

x = "awesome"

def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

# 字串索引(Index)：擷取字串內容

```python
# 透過指標(代表位置)可擷取字串內容，Python為0基底，表示從0開始
word = 'Hello'

# 顯示字串第1個內容
print(word[0])
print(word[1])
print(word[2])
print(word[3])
print(word[4])
```

```python
word = 'Hello'

# 從最後面顯示字串
# -1表最後1個
print(word[-1])
print(word[-2])
print(word[-3])
print(word[-4])
print(word[-5])
```

### 語法：variable_name[start:stop:step] # step表每次進位數，預設為1

```python
# 語法：variable_name[start:stop:step] # step表每次進位數，預設為1

word = 'Hello, World!'

# 從第1個開始，到第2個，亦即顯示0和1位置內容
print(word[0:2]) # stop位置不擷取
```

```python
word = 'Hello, World!'

# 冒號前不指定，代表從頭開始
print(word[:5]) # stop位置不擷取，取0-4
```

```python
word = 'Hello, World!'

# 間隔2，0->2->4
print(word[0:5:2]) # stop位置不擷取

# Hello
# 01234
```

```python
word = 'Hello, World!'

# 未指定結束，代表到最後
print(word[2:]) # 2到結束
```
