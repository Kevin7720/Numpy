## 產生資料

### Series

```python
#----------------------------------------------------------------------
#%%: 產生Series
# 匯入所需套件
import numpy as np  # 匯入成np別名，方便後面使用
import pandas as pd # Pandas一般匯入成pd別名，方便後面使用

# 產生0-9數字, arange([start], stop, [step])
np_a = np.arange(10) # 10表示stop

# 產生Series資料類型
p_s = pd.Series(np_a)

# 顯示於畫面：第一行代表索引，第二行代表對應值
print(p_s)
print("-." * 40) # 分隔線：40個-.

# 從5開始，間隔1，產生到9的數字
np_b = np.arange(5, 10, 1)

# 產生Series資料類型
p_t = pd.Series(np_b)

# 顯示於畫面：第一行代表索引，第二行代表對應值
print(p_t)
```

### 索引(Index)功用

```python
#----------------------------------------------------------------------
#%% 

# 匯入所需套件
import numpy as np
import pandas as pd

# 從5開始，間隔1，產生到9的數字: arange([start], stop, [step])，不含stop
np_b = np.arange(5, 10, 1)

# 產生Series資料類型
p_t = pd.Series(np_b)

# 顯示於畫面：第一行代表索引，第二行代表對應值
print(p_t)
print("-." * 40) # 分隔線：40個-.

# 利用索引取得資料，如未特別設定，將自動從開始累加產生編號
print(p_t[1])  # 表示取得索引1的對應值
```

### 建立索引

```python
#----------------------------------------------------------------------
#%%: 自行指定索引

import pandas as pd

# 產生一串列，內含資料
a = [1, 7, 2]

# 建立Series物件，設定別為x/y/z
myvar = pd.Series(a, index = ["x", "y", "z"])

# 顯示內容
print(myvar)


# 以自訂索引取得值
print(myvar['y'])

# 請自行練習使用其他索引取得值
```

```python
#----------------------------------------------------------------------
#%%: 以字典方式(dictionary)建立索引

# 匯入所需套件
import numpy as np
import pandas as pd

# 以字典方式(dictionary)建立[索引:值]配對關係
dict_test = {"name":"Python",
             "age":10,
             "tel":12345,
             "class":"A班"}

# 產生Series資料類型，並指定索引
t = pd.Series(dict_test)

# 顯示Series
print(t)
print("-." * 40) # 分隔線：40個-.

# 利用索引取得資料
print(t['tel'])  # 表示取得索引B的對應值 
```

```python
#----------------------------------------------------------------------
#%%: 以字典方式(dictionary)建立索引

# 索引(index)和值(values)配對關係: 直接使用index與value取得資料

# 匯入所需套件
import numpy as np
import pandas as pd

# 以字典方式(dictionary)建立索引
dict_test = {"name":"Python",
             "age":10,
             "tel":12345,
             "class":"A班"}

# 產生Series資料類型，並指定索引
t = pd.Series(dict_test)

# 顯示Series
print('index內容 %s ' % t.index) # 顯示索引值
print("-." * 40) # 分隔線：40個-.

print('values內容 %s ' % t.values) # 顯示對應值
```

```python
#----------------------------------------------------------------------
#%%: 

# 匯入所需套件
import numpy as np
import pandas as pd
import string

# 產生0-9數字, 語法：arange([start], stop, [step])
np_a = np.arange(10) # 10表示stop，不包含10，所以產生0-9共10個數字

# 顯示索引值，利用內建英文大寫字母，從第1個開始，共10個
# 請試試string.ascii_lowercase[:10]
print(string.ascii_uppercase[:10])
print("-." * 40) # 分隔線：40個-.

# 產生Series資料類型，並指定索引(以串列方式指定)
p_s = pd.Series(np_a, index = list(string.ascii_uppercase[:10])) # 索引為大寫,string.ascii_uppercase為事先定義好的內容，取第0-9個

# 顯示於畫面：第一行代表索引，第二行代表對應值
print(p_s)
print("-." * 40) # 分隔線：40個-.

# 利用索引取得資料
print(p_s['B'])  # 表示取得索引B的對應值
```

```python
#----------------------------------------------------------------------
#%%

# 匯入pandas套件
import pandas as pd         # industry standard name of pd when importing
import random               # 亂數產生器

# 確保每次產生的亂數皆相同
random.seed(3)   # generate same random numbers every time

# 產生串列
names = ["Jess", "Jordan", "Sandy", "Ted", "Barney", "Tyler", "Rebecca"]

# 顯示names所包含元素個數
print(len(names))

# 加上range函數
print(range(len(names))) 
print("-." * 40) # 分隔線：40個-.

# 在18-35間產生7個數字
ages = [ random.randint(18, 35) for x in range(len(names))] 

# 產生字典資料類型(disctionary)
people = { "names" : names, "ages" : ages }

# 顯示內容
print(people)
print("-." * 40) # 分隔線：40個-.

# 轉換成pandas.DataFrame類型
df = pd.DataFrame.from_dict(people)
#------------------------------------------------------------------------

# 顯示內容
print(df)
```

```python
#----------------------------------------------------------------------
#%%: 6.6 Joins: 連結不同DataFrame成新資料

# inner join

# 匯入pandas套件
import pandas as pd         # industry standard name of pd when importing
import random               # 亂數產生器

# 確保每次產生的亂數皆相同
random.seed(3)   # generate same random numbers every time

# 產生串列
names = ["Jess", "Jordan", "Sandy", "Ted", "Barney", "Tyler", "Rebecca"]

# 在18-35間產生7個數字
ages = [random.randint(18, 35) for x in range(len(names))] 

# 產生字典資料類型(disctionary)
people = {"names" : names, "ages" : ages}

# 轉換成pandas.DataFrame類型
df = pd.DataFrame.from_dict(people)

#------------------------------------------------------------------------
# 在0-9之間產生7個數字
tenure = [ random.randint(0, 10) for x in range(len(df))] 

# 顯示7個數字內容
print(tenure)
print("-." * 40) # 分隔線：40個-.

# 增加一個tenure欄位
df["tenure"] = tenure    

#------------------------------------------------------------------------
# 定義一個函數，可傳入年齡，<21歲回傳'Teenager'，否則回傳'Adult'
def ageGroup(age):
    return "Teenager" if age < 21 else "Adult"

# 套用apply函數，將年齡當參數傳入，並將回傳結果指定給一個新欄位age_group
df["age_group"] = df["ages"].apply(ageGroup)
#------------------------------------------------------------------------

# 產生新字典
ratings = {
    "names": ["Jess", "Tyler", "Ted"],
    "ratings": [10, 9, 6]
}

# 產生新DataFrame
ratings = df.from_dict(ratings)

# 顯示內容
print(ratings.head())
print("-." * 40) # 分隔線：40個-.

# 先顯示內容，第一欄(行)代表索引(index)
print(df)
print("-." * 40) # 分隔線：40個-.

# names為兩個DataFrame相同的欄位，利用names將兩個DataFrame串起
matched_ratings = df.merge(ratings, on = "names", how = "inner") # inner表示兩個DataFrame都須有才能串接(交集)

print(matched_ratings.head())
```
